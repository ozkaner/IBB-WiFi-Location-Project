```python
#Importing necessary libraries
import pandas as pd 
import numpy as np
import folium
from folium.plugins import MarkerCluster, FeatureGroupSubGroup
```

```python
#Reading wifi location data
data =pd.read_csv("ibb_wifi_location.csv")
```

# Observing data

```python
data.head()
```

```python
data.info()
```

```python
data = data.drop(["location_code"],axis=1)
```

```python
data.replace(float(0), np.NaN,inplace = True)
```

```python
data = data.dropna()
```

```python
m = folium.Map(location=[41.015137, 28.979530], zoom_start=11)
```

```python
ilceler = data["location"].unique()
veriler = [data[data["location"]==ilce][['latitude','longitude']].values.tolist() for ilce in ilceler]
tabakalar = [folium.FeatureGroup(name=ilce,show=False) for ilce in ilceler]
for i, ilce in enumerate(ilceler):
    marker_cluster = MarkerCluster(data=veriler[i]).add_to(tabakalar[i])
    for point in range(len(veriler[i])):
        folium.Marker(location=veriler[i][point],
                      icon=folium.Icon(color='blue', icon='wifi', prefix='fa')).add_to(marker_cluster)
for tabaka in tabakalar:
    m.add_child(tabaka)
folium.LayerControl().add_to(m)
```
